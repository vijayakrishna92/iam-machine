importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

import dataset

```python
dt = pd.read_csv("Salary_Data.csv")
x = dt.iloc[:,:-1].values
y = dt.iloc[:,-1].values
```

split datase into training set and test set

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state =0)
```


training simple linear regression model on training set

```python
from sklearn.linear_model import LinearRegression
rig = LinearRegression()
rig.fit(X_train,y_train)
```

predicting test set result

```python
y_pred = rig.predict(X_test)
```

visualising training set result

```python
plt.scatter(X_train, y_train, color= 'red')
plt.plot(X_train,rig.predict(X_train), color ='blue')
plt.title('salary vs experience')
plt.xlabel('y of exp')
plt.ylabel('sal')
plt.show()
```

visualising test set result

```python
plt.scatter(X_test, y_test, color= 'red')
plt.plot(X_train,rig.predict(X_train), color ='blue')
plt.title('salary vs experience')
plt.xlabel('y of exp')
plt.ylabel('sal')
plt.show()
```
